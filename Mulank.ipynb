```python
def get_mulank (dob):
  digits=""
  i=0
  while i<len(dob):
    if dob[i].isdigit():
      digits+=dob[i]
    i+=1

  s=0
  j=0
  while j<len(digits):
    s+=int(digits[j])
    j+=1

  while s>=10:
    temp=0
    n=s
    while n>0:
      temp+=n%10
      n//=10
    s=temp

  return s



personaliti={ 1: { "positive": "Natural leader, confident, ambitious, independent thinker.", "negative": "Can be egoistic, impatient, dominating." },
              2: { "positive": "Diplomatic, cooperative, emotional intelligence strong.", "negative": "Over-sensitive, dependent, indecisive." },
              3: { "positive": "Creative, expressive, social, optimistic.", "negative": "Scattered focus, dramatic, inconsistent." },
              4: { "positive": "Practical, disciplined, hardworking, loyal.", "negative": "Rigid, stubborn, resistant to change." },
              5: { "positive": "Adventurous, energetic, loves freedom, adaptable.", "negative": "Restless, impulsive, lacks consistency." },
              6: { "positive": "Responsible, caring, family-oriented, balanced.", "negative": "Overprotective, controlling, self-sacrificing." },
              7: { "positive": "Analytical, spiritual, intelligent, deep thinker.", "negative": "Overthinking, isolated, emotionally distant." },
              8: { "positive": "Ambitious, business-minded, powerful personality.", "negative": "Materialistic, workaholic, aggressive." },
              9: { "positive": "Compassionate, humanitarian, generous, wise.", "negative": "Emotionally intense, moody, unrealistic." } }


while True:
  print("=====================Numerological Personality App=====================")
  name=input("Enter Your Nme :")
  dob=input("Enter Your Birth Date (DD MM YYYY)").strip()

  digits_only=''.join(c for c in dob if c.isdigit())
  if len(digits_only)<8:
    print("Enter Correct Date of Birth !!!!!")
    print()
  else:
    number=get_mulank(dob)
    traits=personaliti.get(number,{"positive":"Traits not available","negative":"traits not avaioable"})
    print()
    print(f"Hello {name}\n")
    print(f"Your Mulank is : {number}\n")
    print("positive traits :" ,traits["positive"])
    print("negative traits : ",traits["negative"])
    print("====================THANK YOU====================")

```
